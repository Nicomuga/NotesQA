```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('CSVs/analisis_implementacion_ok.csv')

# Perform exploratory analysis
# For example, you can print the first few rows of the DataFrame
# print(df.head())

columnas = df.columns.tolist()
print(columnas)

print(df)

# columnas.remove('columna1')
# columnas.remove('columna2')

# Find the last month in the DataFrame
last_month = df['mes'].max()

# Check if the last month is 12
if last_month == 12:
    # Find the penultimate month
    penultimate_month = df['mes'].nlargest(2).iloc[-1]
    
    # Filter the DataFrame to keep only the rows where remu_cot is not 0 in the last month or penultimate month
    df = df[~((df['mes'] == last_month) | (df['mes'] == penultimate_month)) & (df['remu_cot'] != 0)]
else:
    # Filter the DataFrame to keep only the rows where remu_cot is not 0 in the last month
    df = df[~((df['mes'] == last_month) & (df['remu_cot'] == 0))]

# Verify the changes
print(df)
```
