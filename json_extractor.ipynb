```python
import os
import json

# Parámetros
carpeta_entrada = r"C:\Users\esaez\NotesQA\json_entrada_25_11"  # Ruta de la carpeta con los archivos .json
carpeta_salida = r"C:\Users\esaez\NotesQA\jsons_salida"    # Ruta de la carpeta de salida
lista_folios = ['2024423277','2024423231','2024423251','2024423111','2024416886','2024418661',
'2024418773','2024412236','2024412304']        # Lista de Foliorecepcion a buscar
# lista_folios_str = [str(folio) for folio in lista_folios]
# Crear la carpeta de salida si no existe
os.makedirs(carpeta_salida, exist_ok=True)

# Función para buscar folios y extraer información
def buscar_folios_y_extraer(carpeta_entrada, lista_folios):
    datos_filtrados = []
    for archivo in os.listdir(carpeta_entrada):
        print(archivo)
        if archivo.endswith(".json"):
            ruta_archivo = os.path.join(carpeta_entrada, archivo)
            with open(ruta_archivo, "r", encoding="utf-8") as f:
                datos = json.load(f)
                print('generating large json for search')
                # Filtrar datos por Foliorecepcion
                for registro in datos:
                    if registro.get("Foliorecepcion") in lista_folios:
                        datos_filtrados.append(registro)
    return datos_filtrados

# Extraer información y guardar en un archivo JSON
datos_filtrados = buscar_folios_y_extraer(carpeta_entrada, lista_folios)
if datos_filtrados:
    archivo_salida = os.path.join(carpeta_salida, "datos_filtrados_25_11.json")
    with open(archivo_salida, "w", encoding="utf-8") as f:
        json.dump(datos_filtrados, f, indent=4, ensure_ascii=False)
    print(f"Archivo generado correctamente en: {archivo_salida}")
    print(datos_filtrados)
else:
    print("No se encontraron coincidencias con los folios proporcionados.")
```
