```python
import pandas as pd

# Crear un DataFrame a partir de los datos CSV proporcionados
data = pd.read_csv("CSV/_select_pep_descripcion_Date_fecha_visacion_as_fecha_count_lm_fo_202404101550analisisDevolucionproceso.csv")

# Modificar la columna "fecha" para que sea de tipo fecha
data["fecha"] = pd.to_datetime(data["fecha"], format='%Y-%m-%d')

# Crear un DataFrame con los datos modificados

df = pd.DataFrame(data)
df
```

```python

# Modificar la columna "fecha" para mostrar solo los meses
df["fecha"] = pd.to_datetime(df["fecha"]).dt.to_period('M')




# Mostrar el DataFrame resultante
print(df)
```

```python
import matplotlib.pyplot as plt

# Agrupar los datos por mes y descripción, contar la cantidad de cada descripción por mes
df_grouped = df.groupby(["fecha", "descripcion"]).sum().reset_index()

df_grouped["fecha"] = df_grouped["fecha"].dt.to_timestamp()

df_grouped['suma_mes'] = df.groupby(df['fecha'])['cuenta'].transform('sum')
df_grouped['porcentaje'] = df_grouped['cuenta'] / df_grouped['suma_mes'] * 100
#df_grouped["porcentaje"] = df_grouped.apply(lambda x: (x["cuenta"] / total_cases[x["descripcion"]]) * 100, axis=1)

df_grouped.head(50)


```

Grafico de lineas

```python
import numpy as np

colors = plt.cm.tab20(np.linspace(0, 1, len(df_grouped["descripcion"].unique())))

# Graficar los resultados
plt.figure(figsize=(12, 6))
for i, desc in enumerate(df_grouped["descripcion"].unique()):
    data_desc = df_grouped[df_grouped["descripcion"] == desc]
    plt.plot(data_desc["fecha"], data_desc["porcentaje"], marker='o', label=desc, color=colors[i])
    
plt.title('Análisis de Descripciones por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Colocar la leyenda fuera del gráfico
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Graficar los resultados en un gráfico de barras al 100%
plt.figure(figsize=(12, 6))

# Plot de barras
for desc in df_grouped["descripcion"].unique():
    data_desc = df_grouped[df_grouped["descripcion"] == desc]
    plt.bar(data_desc["fecha"], data_desc["porcentaje"], label=desc)

plt.title('Análisis de Descripciones - Porcentaje de casos')
plt.xlabel('Mes')
plt.ylabel('Porcentaje de casos (%)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Crear recuadro de descripciones aparte del gráfico
plt.figure(figsize=(8, 4))
for i, desc in enumerate(df_grouped["descripcion"].unique()):
    plt.text(0, i, desc, fontsize=10, ha='right')
plt.axis('off')

plt.show()
```
