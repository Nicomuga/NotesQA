```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('CSVs/cuadratura_dv_proceso_MAYO.csv')

# Print the DataFrame
print(df)
```

```python
import pandas as pd

# Transform fecha_recepcion to datetime format
df['fecha_recepcion'] = pd.to_datetime(df['fecha_recepcion'])

# Subtract 4 hours from all the records
df['fecha_recepcion_adjusted'] = df['fecha_recepcion'] - pd.Timedelta(hours=4)

# Print the updated DataFrame
filtered_df = df.loc[df['fecha_recepcion'].dt.date == pd.to_datetime('2024-04-30').date()]
filtered_df
```

```python

df['contract_date'] = df['fecha_recepcion_adjusted']
df.loc[df['contract_date'].dt.hour >= 17, 'contract_date'] += pd.Timedelta(days=1)
df.loc[df['contract_date'].dt.dayofweek == 5, 'contract_date'] += pd.Timedelta(days=2)
df['contract_date'] = df['contract_date'].dt.date

df
```

```python
pivot_table = df.pivot_table(index='contract_date', columns='estado_final_lm', aggfunc='size', fill_value=0)
pivot_table = pivot_table.apply(lambda x: x / x.sum(), axis=1)

pivot_table

```

```python
import pandas as pd

# Generar la tabla pivote con conteo
pivot_table_2 = pd.pivot_table(df, values='folio_recepcion', index='contract_date', columns='estado_final_lm', aggfunc='count', fill_value=0)

# Agregar una columna con el conteo final por fecha
pivot_table_2['conteo_final'] = pivot_table_2.sum(axis=1)

# Mostrar la tabla pivote
print(pivot_table_2)
```
