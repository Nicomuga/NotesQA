```python
import boto3
import os

# Configurar el perfil de AWS
session = boto3.Session(profile_name='msys-consalud')

# Inicializar el cliente de S3
s3 = session.client('s3')

# URI de S3 del bucket y el prefijo del path
s3_uri = 's3://consaludsil/prod/publicacion_documentos_procesados/'

# Extraer el nombre del bucket y el prefijo del path del URI de S3
bucket_name = s3_uri.split('/')[2]
prefijo_path = '/'.join(s3_uri.split('/')[3:])

# Lista de nombres de archivos a descargar (sin el ' ok' final y con extensi√≥n .json)
archivos_a_descargar = [
    '2024204538.json',
    '2024204928.json',
    '2024204973.json',
    '2024205295.json',
    '2024207139.json',
    '2024207180.json',
    '2024207190.json',
    '2024207217.json',
    '2024207231.json',
    '2024207427.json',
    '2024207470.json',
    '2024207521.json',
    '2024207575.json',
    '2024207760.json',
    '2024207804.json',
    '2024207827.json',
    '2024207835.json',
    '2024207956.json',
    '2024207980.json',
    '2024210021.json',
    '2024210070.json',
    '2024210111.json',
    '2024210141.json',
    '2024496091.json',
    '2024496211.json',
    '2024496555.json',
    '2024003387.json',
    '2024094091.json',
    '2024204037.json',
    '2024207363.json'
]

# Directorio de destino
directorio_destino = 'Folios_S3\Folios_procesados_06-06'

# Crear el directorio si no existe
os.makedirs(directorio_destino, exist_ok=True)

# Descargar los archivos
for archivo in archivos_a_descargar:
    # Construir el path completo del archivo en S3
    file_name = os.path.join(prefijo_path, archivo)
    # Construir la ruta completa del archivo local
    local_file_name = os.path.join(directorio_destino, archivo)
    # Descargar el archivo
    s3.download_file(bucket_name, file_name, local_file_name)
    print(f"Descargado {file_name} como {local_file_name}")

print("Descarga completa")
```
